```python
# Day 4 - Risk summary dashboard + creatinine
patients_ages = [45, 52, 67, 71, 38, 60]
patients_hba1c = [9.2, 7.8, 11.1, 8.4, 10.3, 6.9]
patients_abi = [0.72, 0.95, 0.65, 0.88, 0.61, 1.02]
patients_creatinine = [1.8, 0.9, 2.4, 1.1, 2.9, 0.8]
print("Multi-Variable Patients Risk Assessment:")
print("---")
high_risk_count = 0
moderate_risk_count = 0
low_risk_count = 0
for i in range(len(patients_ages)):
    patient_label = f"patient{i+1}"
    high_hba1c = patients_hba1c[i] > 8.0
    low_abi = patients_abi[i] < 0.9
    high_creatinine = patients_creatinine[i] > 1.2
    risk_score = sum([high_hba1c, low_abi, high_creatinine])
    if risk_score == 3:
        status = "HIGH RISK"
        high_risk_count += 1
    elif risk_score == 2:
        status = "MODERATE RISK"
        moderate_risk_count += 1
    else:
        status = "Low risk"
        low_risk_count += 1
    print(f"{patient_label} | Age: {patients_ages[i]} | HbA1c: {patients_hba1c[i]}% | ABI: {patients_abi[i]} | Creatinine: {patients_creatinine[i]} | STATUS: {status}") 
print("---")
print("Risk Summary Dashboard:")
print("---")
print(f"HIGH RISK patients:     {high_risk_count}")
print(f"MODERATE RISK patients: {moderate_risk_count}")
print(f"Low risk patients:      {low_risk_count}")
print(f"Total patients reviewed: {len(patients_ages)}")
print(f"Action required: {high_risk_count} patients need urgent vascular review")
```

```python
# Day 5 - building a reusable risk assessment function
def assess_patient(name, age, hba1c, abi, creatinine):
    high_hba1c = hba1c > 8.0
    low_abi = abi < 0.9
    high_creatinine = creatinine > 1.2
    risk_score = sum([high_hba1c, low_abi, high_creatinine])
    if risk_score == 3:
        status = "HIGH RISK"
    elif risk_score == 2:
        status = "MODERATE RISK"
    else:
        status = "Low risk"
    print(f"{name} | Age: {age} | HbA1c: {hba1c}% | ABI: {abi} | Creatinine: {creatinine} | SCORE: {risk_score}/3 | STATUS: {status}")
    return status
print("--- Patient Risk Assessments ---")
assess_patient("Patient 1", 45, 9.2, 0.72, 1.8)
assess_patient("Patient 2", 52, 7.8, 0.95, 0.9)
assess_patient("Patient 3", 67, 11.1, 0.65, 2.4)
assess_patient("Patient 4", 71, 8.4, 0.88, 1.1)
assess_patient("Patient 5", 38, 10.3, 0.61, 2.9)
assess_patient("Patient 6", 60, 6.9, 1.02, 0.8)
print("---")
print("\n--- Testing with a new patient---")
assess_patient("Dr Ogunlade's Patient", 58, 9.8, 0.71, 1.6)
```

```python
# Day 6 - Batch assessment function
# Self-contained - no dependency on previous cells

def assess_ward(patient_list):
    print("--- Ward Risk Assessment ---")
    print(f"{'Patient':<20} {'Age':<6} {'HbA1c':<8} {'ABI':<6} {'Creatinine':<12} {'Score':<8} {'Status'}")
    print("-" * 80)
    
    high = 0
    moderate = 0
    low = 0
    
    for patient in patient_list:
        name = patient["name"]
        age = patient["age"]
        hba1c = patient["hba1c"]
        abi = patient["abi"]
        creatinine = patient["creatinine"]
        
        risk_score = sum([hba1c > 8.0, abi < 0.9, creatinine > 1.2])
        
        if risk_score == 3:
            status = "HIGH RISK"
            high += 1
        elif risk_score == 2:
            status = "MODERATE RISK"
            moderate += 1
        else:
            status = "Low risk"
            low += 1
        
        print(f"{name:<20} {age:<6} {hba1c:<8} {abi:<6} {creatinine:<12} {risk_score}/3{'':<4} {status}")
    
    print("-" * 80)
    print(f"HIGH RISK:     {high}")
    print(f"MODERATE RISK: {moderate}")
    print(f"Low risk:      {low}")
    print(f"Urgent referrals needed: {high}")

ward_patients = [
    {"name": "Emmanuel A.", "age": 45, "hba1c": 9.2, "abi": 0.72, "creatinine": 1.8},
    {"name": "Adebayo T.", "age": 52, "hba1c": 7.8, "abi": 0.95, "creatinine": 0.9},
    {"name": "Chukwu F.", "age": 67, "hba1c": 11.1, "abi": 0.65, "creatinine": 2.4},
    {"name": "Fatima M.", "age": 71, "hba1c": 8.4, "abi": 0.88, "creatinine": 1.1},
    {"name": "Tunde B.", "age": 38, "hba1c": 10.3, "abi": 0.61, "creatinine": 2.9},
    {"name": "Ngozi K.", "age": 60, "hba1c": 6.9, "abi": 1.02, "creatinine": 0.8},
    {"name": "Isreal Z.", "age": 75, "hba1c": 8.6, "abi": 0.81, "creatinine": 1.1},
]

assess_ward(ward_patients)
```

```python
# Day 7 - Exporting Clinical report to CSV
import csv
from datetime import datetime

def assess_and_export(patient_list, filename="vascular_risk_report.csv"):

    report_date = datetime.now() .strftime("%Y-%m-%d %H:%M")
    results = []

    high = 0
    moderate = 0
    low = 0

    for patient in patient_list:
        name = patient["name"]
        age = patient["age"]
        hba1c = patient["hba1c"]
        abi = patient["abi"]
        creatinine = patient["creatinine"]

        high_hba1c = hba1c > 8.0
        low_abi =abi < 0.9
        high_creatinine = creatinine > 1.2
        risk_score = sum([high_hba1c, low_abi, high_creatinine])

        if risk_score == 3:
            status = "HIGH RISK"
            action = "Urgent vascular referral"
            high += 1
        elif risk_score == 2:
            status = "MODERATE RISK"
            action = "Aggressive optimisation + monthly review"
            moderate += 1
        else:
            status = "Low risk"
            action = "Standard follow-up"
            low += 1

        results.append({
            "Patient": name,
            "Age": age,
            "HbA1c (%)": hba1c,
            "ABI": abi,
            "Creatnine": creatinine,
            "Risk Score": f"{risk_score}/3",
            "Status": status,
            "Recommended Action": action,
            "Report Date": report_date
        })
        
    with open(filename, "w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=results[0].keys())
        writer.writeheader()
        writer.writerows(results)

    print(f"Report generated: {filename}")
    print(f"Date: {report_date}")
    print(f"Total patients: {len(patient_list)}")
    print(f"---")
    print(f"HIGH RISK:     {high} - Urgemt vascular referral")
    print(f"MODERATE RISK: {moderate} - Aggressive optimisation")
    print(f"Low risk:      {low} - Standard follow-up")
    print(f"---")
    print(f"File saved to your vascular-research folder.")

ward_patients = [
    {"name": "Emmanuel A.", "age": 45, "hba1c": 9.2, "abi": 0.72, "creatinine": 1.8},
    {"name": "Adebayo T.", "age": 52, "hba1c": 7.8, "abi": 0.95, "creatinine": 0.9},
    {"name": "Chukwu F.", "age": 67, "hba1c": 11.1, "abi": 0.65, "creatinine": 2.4},
    {"name": "Fatima M.", "age": 71, "hba1c": 8.4, "abi": 0.88, "creatinine": 1.1},
    {"name": "Tunde B.", "age": 38, "hba1c": 10.3, "abi": 0.61, "creatinine": 2.9},
    {"name": "Ngozi K.", "age": 60, "hba1c": 6.9, "abi": 1.02, "creatinine": 0.8},
    {"name": "Isreal Z.", "age": 75, "hba1c": 8.6, "abi": 0.81, "creatinine": 1.1},
]

assess_and_export(ward_patients)


```
