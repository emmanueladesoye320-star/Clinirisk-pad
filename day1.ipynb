```python
# Day 5 - building a reusable risk assessment function
def assess_patient(name, age, hba1c, abi, creatinine):
    high_hba1c = hba1c > 8.0
    low_abi = abi < 0.9
    high_creatinine = creatinine > 1.2
    risk_score = sum([high_hba1c, low_abi, high_creatinine])
    if risk_score == 3:
        status = "HIGH RISK"
    elif risk_score == 2:
        status = "MODERATE RISK"
    else:
        status = "Low risk"
    print(f"{name} | Age: {age} | HbA1c: {hba1c}% | ABI: {abi} | Creatinine: {creatinine} | SCORE: {risk_score}/3 | STATUS: {status}")
    return status
print("--- Patient Risk Assessments ---")
assess_patient("Patient 1", 45, 9.2, 0.72, 1.8)
assess_patient("Patient 2", 52, 7.8, 0.95, 0.9)
assess_patient("Patient 3", 67, 11.1, 0.65, 2.4)
assess_patient("Patient 4", 71, 8.4, 0.88, 1.1)
assess_patient("Patient 5", 38, 10.3, 0.61, 2.9)
assess_patient("Patient 6", 60, 6.9, 1.02, 0.8)
print("---")
print("\n--- Testing with a new patient---")
assess_patient("Dr Ogunlade's Patient", 58, 9.8, 0.71, 1.6)
```

```python
# Day 6 - Batch assessment function
# Self-contained - no dependency on previous cells

def assess_ward(patient_list):
    print("--- Ward Risk Assessment ---")
    print(f"{'Patient':<20} {'Age':<6} {'HbA1c':<8} {'ABI':<6} {'Creatinine':<12} {'Score':<8} {'Status'}")
    print("-" * 80)
    
    high = 0
    moderate = 0
    low = 0
    
    for patient in patient_list:
        name = patient["name"]
        age = patient["age"]
        hba1c = patient["hba1c"]
        abi = patient["abi"]
        creatinine = patient["creatinine"]
        
        risk_score = sum([hba1c > 8.0, abi < 0.9, creatinine > 1.2])
        
        if risk_score == 3:
            status = "HIGH RISK"
            high += 1
        elif risk_score == 2:
            status = "MODERATE RISK"
            moderate += 1
        else:
            status = "Low risk"
            low += 1
        
        print(f"{name:<20} {age:<6} {hba1c:<8} {abi:<6} {creatinine:<12} {risk_score}/3{'':<4} {status}")
    
    print("-" * 80)
    print(f"HIGH RISK:     {high}")
    print(f"MODERATE RISK: {moderate}")
    print(f"Low risk:      {low}")
    print(f"Urgent referrals needed: {high}")

ward_patients = [
    {"name": "Emmanuel A.", "age": 45, "hba1c": 9.2, "abi": 0.72, "creatinine": 1.8},
    {"name": "Adebayo T.", "age": 52, "hba1c": 7.8, "abi": 0.95, "creatinine": 0.9},
    {"name": "Chukwu F.", "age": 67, "hba1c": 11.1, "abi": 0.65, "creatinine": 2.4},
    {"name": "Fatima M.", "age": 71, "hba1c": 8.4, "abi": 0.88, "creatinine": 1.1},
    {"name": "Tunde B.", "age": 38, "hba1c": 10.3, "abi": 0.61, "creatinine": 2.9},
    {"name": "Ngozi K.", "age": 60, "hba1c": 6.9, "abi": 1.02, "creatinine": 0.8},
    {"name": "Isreal Z.", "age": 75, "hba1c": 8.6, "abi": 0.81, "creatinine": 1.1},
]

assess_ward(ward_patients)
```

```python
# Day 7 - Exporting Clinical report to CSV
import csv
from datetime import datetime

def assess_and_export(patient_list, filename="vascular_risk_report.csv"):

    report_date = datetime.now() .strftime("%Y-%m-%d %H:%M")
    results = []

    high = 0
    moderate = 0
    low = 0

    for patient in patient_list:
        name = patient["name"]
        age = patient["age"]
        hba1c = patient["hba1c"]
        abi = patient["abi"]
        creatinine = patient["creatinine"]

        high_hba1c = hba1c > 8.0
        low_abi =abi < 0.9
        high_creatinine = creatinine > 1.2
        risk_score = sum([high_hba1c, low_abi, high_creatinine])

        if risk_score == 3:
            status = "HIGH RISK"
            action = "Urgent vascular referral"
            high += 1
        elif risk_score == 2:
            status = "MODERATE RISK"
            action = "Aggressive optimisation + monthly review"
            moderate += 1
        else:
            status = "Low risk"
            action = "Standard follow-up"
            low += 1

        results.append({
            "Patient": name,
            "Age": age,
            "HbA1c (%)": hba1c,
            "ABI": abi,
            "Creatnine": creatinine,
            "Risk Score": f"{risk_score}/3",
            "Status": status,
            "Recommended Action": action,
            "Report Date": report_date
        })
        
    with open(filename, "w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=results[0].keys())
        writer.writeheader()
        writer.writerows(results)

    print(f"Report generated: {filename}")
    print(f"Date: {report_date}")
    print(f"Total patients: {len(patient_list)}")
    print(f"---")
    print(f"HIGH RISK:     {high} - Urgemt vascular referral")
    print(f"MODERATE RISK: {moderate} - Aggressive optimisation")
    print(f"Low risk:      {low} - Standard follow-up")
    print(f"---")
    print(f"File saved to your vascular-research folder.")

ward_patients = [
    {"name": "Emmanuel A.", "age": 45, "hba1c": 9.2, "abi": 0.72, "creatinine": 1.8},
    {"name": "Adebayo T.", "age": 52, "hba1c": 7.8, "abi": 0.95, "creatinine": 0.9},
    {"name": "Chukwu F.", "age": 67, "hba1c": 11.1, "abi": 0.65, "creatinine": 2.4},
    {"name": "Fatima M.", "age": 71, "hba1c": 8.4, "abi": 0.88, "creatinine": 1.1},
    {"name": "Tunde B.", "age": 38, "hba1c": 10.3, "abi": 0.61, "creatinine": 2.9},
    {"name": "Ngozi K.", "age": 60, "hba1c": 6.9, "abi": 1.02, "creatinine": 0.8},
    {"name": "Isreal Z.", "age": 75, "hba1c": 8.6, "abi": 0.81, "creatinine": 1.1},
]

assess_and_export(ward_patients)


```

```python
# Day 8 - Data ingestion pipeline
# Tool now accepts any real CSV dataset automatically

import pandas as pd
import os
from datetime import datetime

def load_and_assess(filepath,
                    age_col,
                    hba1c_col,
                    abi_col,
                    creatinine_col,
                    name_col=None):
    
    print(f"Loading dataset: {filepath}")
    df = pd.read_csv(filepath)
    print(f"Dataset loaded - {len(df)} records found.")
    print("---")

    # Clean missing values
    required = [age_col, hba1c_col, abi_col, creatinine_col]
    df = df.dropna(subset=required)
    print(f"Records after cleaning: {len(df)}")
    print("---")

    results = []
    high = moderate = low = 0

    for idx, row in df.iterrows():
        name = str(row[name_col]) if name_col else f"Patient {idx+1}"
        age = row[age_col]
        hba1c = row[hba1c_col]
        abi = row[abi_col]
        creatinine = row[creatinine_col]

        risk_score = sum([hba1c > 8.0, abi < 0.9, creatinine > 1.2])
        
        if risk_score == 3:
            status = "HIGH RISK"
            action = "Urgent vascular referral"
            high += 1
        elif risk_score == 2:
            status = "MODERATE RISK"
            action = "Aggressive optimisation + monthly review"
            moderate += 1
        else:
            status = "Low risk"
            action = "Standard follow-up"
            low += 1
        
        results.append({
            "Patient": name,
            "Age": age,
            "HbA1c (%)": hba1c,
            "ABI": abi,
            "Creatinine": creatinine,
            "Risk Score": f"{risk_score} of 3",
            "Status": status,
            "Action": action,
            "Date": datetime.now().strftime("%Y-%m-%d")
        })
    
    # Print summary
    print("CLTI RISK ASSESSMENT SUMMARY")
    print("=" * 40)
    print(f"Total patients assessed: {len(results)}")
    print(f"HIGH RISK:     {high} — Urgent vascular referral")
    print(f"MODERATE RISK: {moderate} — Aggressive optimisation")
    print(f"Low risk:      {low} — Standard follow-up")
    print("=" * 40)
    
    # Export report
    output_name = f"ClinRisk_Report_{datetime.now().strftime('%Y%m%d_%H%M')}.csv"
    pd.DataFrame(results).to_csv(output_name, index=False)
    print(f"Report exported: {output_name}")
    
    return pd.DataFrame(results)

# Test with your existing ward data
test_data = {
    "Name":       ["Emmanuel A.", "Adebayo T.", "Chukwu F.", 
                   "Fatima M.",   "Tunde B.",   "Ngozi K.", "Isreal Z."],
    "Age":        [45, 52, 67, 71, 38, 60, 75],
    "HbA1c":      [9.2, 7.8, 11.1, 8.4, 10.3, 6.9, 8.6],
    "ABI":        [0.72, 0.95, 0.65, 0.88, 0.61, 1.02, 0.81],
    "Creatinine": [1.8, 0.9, 2.4, 1.1, 2.9, 0.8, 1.1]
}

# Save test data as CSV first
pd.DataFrame(test_data).to_csv("test_ward.csv", index=False)
print("Test dataset created: test_ward.csv")
print("---")

# Now run the full pipeline
results = load_and_assess(
    filepath="test_ward.csv",
    name_col="Name",
    age_col="Age",
    hba1c_col="HbA1c",
    abi_col="ABI",
    creatinine_col="Creatinine"
)
```

```python
# Day 9 - First visualisation
# Risk distribution chart across patient cohort

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd
from datetime import datetime

# Patient data
patients = [
    {"name": "Emmanuel A.", "age": 45, "hba1c": 9.2,  "abi": 0.72, "creatinine": 1.8},
    {"name": "Adebayo T.",  "age": 52, "hba1c": 7.8,  "abi": 0.95, "creatinine": 0.9},
    {"name": "Chukwu F.",   "age": 67, "hba1c": 11.1, "abi": 0.65, "creatinine": 2.4},
    {"name": "Fatima M.",   "age": 71, "hba1c": 8.4,  "abi": 0.88, "creatinine": 1.1},
    {"name": "Tunde B.",    "age": 38, "hba1c": 10.3, "abi": 0.61, "creatinine": 2.9},
    {"name": "Ngozi K.",    "age": 60, "hba1c": 6.9,  "abi": 1.02, "creatinine": 0.8},
    {"name": "Isreal Z.",   "age": 75, "hba1c": 8.6,  "abi": 0.81, "creatinine": 1.1},
]

# Calculate risk scores
for p in patients:
    p["risk_score"] = sum([
        p["hba1c"] > 8.0,
        p["abi"] < 0.9,
        p["creatinine"] > 1.2
    ])
    if p["risk_score"] == 3:
        p["status"] = "HIGH RISK"
        p["color"] = "#C0392B"
    elif p["risk_score"] == 2:
        p["status"] = "MODERATE RISK"
        p["color"] = "#E67E22"
    else:
        p["status"] = "Low risk"
        p["color"] = "#27AE60"

df = pd.DataFrame(patients)

# ── Figure with 2 subplots ──────────────────────────────────────────────
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
fig.patch.set_facecolor('#0F0F0F')

for ax in [ax1, ax2]:
    ax.set_facecolor('#1A1A1A')
    ax.tick_params(colors='white')
    ax.xaxis.label.set_color('white')
    ax.yaxis.label.set_color('white')
    ax.title.set_color('white')
    for spine in ax.spines.values():
        spine.set_edgecolor('#333333')

# ── Chart 1: Risk score per patient ────────────────────────────────────
bars = ax1.barh(
    df["name"],
    df["risk_score"],
    color=df["color"],
    edgecolor='#0F0F0F',
    height=0.6
)

ax1.set_xlim(0, 3.5)
ax1.set_xlabel("Risk Score (out of 3)", fontsize=11)
ax1.set_title("ClinRisk-PAD — Individual Patient Risk Scores",
              fontsize=12, fontweight='bold', pad=15)
ax1.axvline(x=2, color='#E67E22', linestyle='--',
            alpha=0.5, linewidth=1, label='Moderate threshold')
ax1.axvline(x=3, color='#C0392B', linestyle='--',
            alpha=0.5, linewidth=1, label='High threshold')

for bar, score in zip(bars, df["risk_score"]):
    ax1.text(bar.get_width() + 0.05, bar.get_y() + bar.get_height()/2,
             f'{score}/3', va='center', color='white', fontsize=10)

# ── Chart 2: Risk distribution donut ───────────────────────────────────
counts = df["status"].value_counts()
status_order = ["HIGH RISK", "MODERATE RISK", "Low risk"]
colors_order = ["#C0392B", "#E67E22", "#27AE60"]

labels = []
sizes = []
colors = []
for s, c in zip(status_order, colors_order):
    if s in counts:
        labels.append(s)
        sizes.append(counts[s])
        colors.append(c)

wedges, texts, autotexts = ax2.pie(
    sizes,
    labels=None,
    colors=colors,
    autopct='%1.0f%%',
    startangle=90,
    pctdistance=0.75,
    wedgeprops=dict(width=0.5, edgecolor='#0F0F0F', linewidth=2)
)

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')

legend_patches = [
    mpatches.Patch(color=c, label=f"{l} (n={s})")
    for l, s, c in zip(labels, sizes, colors)
]
ax2.legend(handles=legend_patches, loc='lower center',
           bbox_to_anchor=(0.5, -0.12), ncol=1,
           framealpha=0, labelcolor='white', fontsize=10)

ax2.set_title("Risk Distribution — Ward Cohort",
              fontsize=12, fontweight='bold', pad=15)

# Centre text
ax2.text(0, 0, f"{len(df)}\npatients",
         ha='center', va='center',
         color='white', fontsize=13, fontweight='bold')

# ── Footer ──────────────────────────────────────────────────────────────
fig.text(0.5, 0.01,
         f"ClinRisk-PAD  ·  Generated {datetime.now().strftime('%B %d, %Y')}  ·  "
         f"github.com/emmanueladesoye320-star/Clinirisk-pad",
         ha='center', color='#666666', fontsize=9)

plt.suptitle("ClinRisk-PAD Clinical Risk Assessment Dashboard",
             fontsize=15, fontweight='bold', color='white', y=1.02)

plt.tight_layout()
plt.savefig("ClinRisk_Dashboard.png", dpi=150,
            bbox_inches='tight', facecolor='#0F0F0F')
plt.show()
print("Dashboard saved: ClinRisk_Dashboard.png") 
```
